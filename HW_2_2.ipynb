```python
# # List.py 
# Implementation of Python LIST as dynamic array
# Author: Jagadeesh Vasudevamurthy
# Copyright: Jagadeesh Vasudevamurthy 2021
###########################################################

############################################################
#  All imports here
###########################################################
import ctypes

############################################################
#  List (Java like List implemented as dynamic array)
###########################################################
class List(object):
    def __init__(self):
        self._k = 0 ;
        self._capacity = 8
        self._a = self._allocate(self._capacity)

    ############################################################
    #  All Public functions below
    #  WRITE CODE BELOW
    ###########################################################
    
    def append(self, item):
        if self._k == self._capacity:
            self._grow()
        self._a[self._k] = item
        self._k = self._k + 1
    
    def delete(self, item):
        result = list()
        n = len(self)
        j = 0
        for i in range(n):
            vi = self[i]
            if vi != item:
                result.append(vi)
                j+=1
        self._k = j
        for i in range(len(self)):
            self._a[i] = result[i]
            
    ############################################################
    #  All private functions below
    # WRITE CODE BELOW
    ###########################################################

    ############################################################
    #  Time:O(1)
    #  Space:O(1)
    # Cannot be changed
    ###########################################################
    def _allocate(self,new_cap):
        return (new_cap * ctypes.py_object)()
    
    
    def __len__(self):
        return self._k
    
    def __getitem__(self, index):
        assert(index>=0 and index< self._k)
        return self._a[index]
    
    def __setitem__(self, index, value):
        assert(index>=0 and index< self._k)
        self._a[index] = value
        
    def __contains__(self, item)->'int':
        for i in range(self._k):
            if(self._a[i]==item):
                return i
            return -1
    
    def __str__(self)->'String':
        s = "[ " 
        for i in range(self._k):
            s+=str(self._a[i]) +" "
        s+="]"
        return s
    
    def __delitem__(self, index):
        assert(index>=0 and index< self._k)
        j = index + 1
        for k in range(j, self._k):
            self._a[k-1] = self._a[k]
        self._k = self._k - 1
        
    ############################################################
    #  Time:O(n)
    #  Space:O(n)
    # Cannot be changed
    ###########################################################
    def _grow(self):
        os =  self._capacity
        ns = os * 2
        print("Grow from", os, "to", ns, ". This is not O(1)")
        b = self._allocate(ns) # New bigger array
        
        for k in range(os): # Reference all existing values
            b[k] = self._a[k]    
        self._a = b # Call b as the new bigger array
        self._capacity = ns # Reset the capacity
  
```
