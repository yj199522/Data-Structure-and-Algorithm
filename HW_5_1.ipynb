```python
############################################################
# Recursion.py 
# Author: Jagadeesh Vasudevamurthy
# Copyright: Jagadeesh Vasudevamurthy 2021
###########################################################

############################################################
#  All imports here
###########################################################

############################################################
#  class  test factorial
###########################################################    
class Recursion():
    def __init__(self):
        #Nothing can be added here
        self.noOfCombination = 0
        self.noOfMoves = 0
        pass

    ############################################################
    #          WRITE ALL YOUR PUBLIC FUNCTION BELOW
    ########################################################### 
    def FibI(self, n):
        if n < 2:
            return n
        secondlast = 0
        last = 1
        for i in range(2,n+1):
            ans = secondlast + last
            secondlast = last
            last = ans
        return ans
    
    def FibR(self, n):
        if n < 2:
            return n
        return self.FibR(n-1) + self.FibR(n-2)
    
    def permR(self, n):
        if n < 2:
            return 1
        arr = [0] * (n)
        for i in range(n):
            arr[i] = i
        self.permutationHelper(arr, 0, n)
        return self.noOfCombination
    
    def permutationHelper(self, arr, start, end):
        if start==end:
            self.noOfCombination+=1
            print(self.noOfCombination," :",arr)
            return
        for i in range(start, end):
            arr = self.swap(arr, i, start)
            self.permutationHelper(arr, start+1,end)
            arr = self.swap(arr, i, start)
    
    def swap(self, arr, start, end):
        if start == end:
            return arr
        arr[start], arr[end] = arr[end], arr[start]
        return arr
    
    def thR(self, n):
        self.thRHelper(n, 1, 2, 3)
        return self.noOfMoves
    
    def thRHelper(self, n, start, temp, dest):
        if n:
            self.thRHelper(n - 1, start, dest, temp)
            self.noOfMoves+=1
            print(self.noOfMoves," : ", start, "->", dest)
            self.thRHelper(n - 1, temp, start, dest)
            
    ############################################################
    #          WRITE ALL YOUR PRIVATE FUNCTION BELOW
    ########################################################### 
   
```
