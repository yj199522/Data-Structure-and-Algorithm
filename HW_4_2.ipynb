```python
############################################################
# Recursion.py 
# Author: Jagadeesh Vasudevamurthy
# Copyright: Jagadeesh Vasudevamurthy 2021
###########################################################

############################################################
#  All imports here
###########################################################

############################################################
#  class  test factorial
###########################################################    
class Recursion():
    def __init__(self):
        #Nothing can be added here
        pass
    ############################################################
    #          WRITE ALL YOUR PUBLIC FUNCTION BELOW
    ########################################################### 
    def factI(self, n):
        if n == 0 :
            return 1
        result = 1
        while n !=1:
            result*=n
            n-=1
        return result
    
    def factR(self, n):
        if n == 0 :
            return 1
        return n * self.factR(n-1)

    def printI(self, n, reverse):
        if abs(n) < 10:
            print(n)
        elif reverse:
            print(self.IterativeRev(n, 0))
        else: 
            print(n)
    
    def printR(self, n, reverse):
        if abs(n) < 10:
            print(n)
        elif reverse:
            print(self.recursionRev(n, 0))
        else:
            print(n)

    def IterativeRev(self, n, result):
        while n!=0:
            result = result * 10 + n % 10
            n //= 10
        return result

    def recursionRev(self, n, value):
        if n==0:
            return value
        else:
            return self.recursionRev(n//10, value*10 + n%10)
    
    def reverseI(self, n):
        return self.IterativeRev(n, 0)
    
    def reverseR(self, n):
        return self.recursionRev(n, 0)
    ############################################################
    #          WRITE ALL YOUR PRIVATE FUNCTION BELOW
    ########################################################### 
 
```
