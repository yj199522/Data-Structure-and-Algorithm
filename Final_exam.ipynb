```python
############################################################
# Exam.py 
# 
###########################################################

############################################################
#  All imports here
###########################################################

############################################################
#  class  Alg
###########################################################    
class Alg():
    def __init__(self,a:'python list',ans:'python list',maxv:'list of size 1',work:'list of size 1',show:'bool'):
        self._a = a
        self._ans = ans
        self._maxv = maxv
        self._work = work
        self._show = show
        ## You can have your data structure here
        
        ## Nothing can be changed below
        self._exam() #Everything happens in _exam
        check_result(self._a,self._ans,self._maxv[0]) #your answer is checked here

    ############################################################
    #          Nothing can be changed in _exam
    ########################################################### 
    def _exam(self):
        self._alg()
        check_result(self._a,self._ans,self._maxv[0]) #your answer is checked here

    ############################################################
    #          WRITE CODE BELOW
    #  Populate self._maxv and self._work 
    #  if self._show is True, show steps of the algorithm
    ########################################################### 
    def _alg(self):
        n = len(self._a)
        if self._show:
            print("*************** Solution ***************")
            print()
        if n == 0:
            self._maxv[0] = 0
            if self._show:
                print("Since the list is Empty then the max value is 0")
                print()
        elif n == 1:
            self._work[0]+=1
            self._maxv[0] = self._a[-1]
            self._ans.append(0)
            if self._show:
                print("Since the list contain only one value then the max value is:", self._a[-1])
                print()
        else:
            self._work[0]+=1
            excludeIndex, includeIndex, include, exclude = [], [0], self._a[0], 0
            if self._show:
                print("Interation No: ", self._work[0])
                print("At index 0 the number is", self._a[0])
                print("Since previous sum is less then current sum 0 <", self._a[0] ,". So the new sum till index 0 is", include, ". So new included index list is:", includeIndex)
                print()
            for i in range(1, len(self._a)):
                temp = include
                self._work[0]+=1
                if self._show:
                    print("Interation No: ", self._work[0])
                if include < exclude + self._a[i]:
                    include = exclude + self._a[i]
                    excludeIndex.append(i)
                    tempList = includeIndex[:]
                    includeIndex, excludeIndex = excludeIndex[:], tempList[:]
                    if self._show:
                        print("At index",i,"the number is", self._a[i])
                        print("Since previous sum is less then current sum ", temp, "<", include ,". So the new sum till index",i,"is",  include, ". So new included index list is:", includeIndex)
                        print()
                else:
                    excludeIndex = includeIndex[:]
                    if self._show:
                        print("At index",i,"the number is", self._a[i])
                        print("Since previous sum is greater or equal to then current sum ", include ,">=", exclude + self._a[i] ,". So we are ignoring the value", self._a[i], "in the sum.")
                        print()
                exclude = temp
            if include >= exclude:
                self._maxv[0] = include
                self._ans.extend(includeIndex)
            else:
                self._maxv[0] = exclude
                self._ans.extend(excludeIndex)
        if self._show:
            print("Since we are traversing throught the entired list i.e. the work done is",self._work[0],"that is equal to the length of the list which is", len(self._a),". So the Time Complexity is O(N)")
            print()
            print("Since we are tracking the indexes of the elements in the array, the maximum possible length of the array would be N/2. Therefore, the space complexity is O(N)")
            print()

############################################################
#  NOTHING CAN BE CHANGED BELOW
###########################################################  

############################################################
#  class  Solution
# Nothing can be changed in Solution
###########################################################    
class Solution():
    def rob(self, nums:'Python list') -> 'int':
        #Nothing can be changedhere
        ans = []
        maxv = [0]
        work = [0]
        t = Alg(nums,ans,maxv,work,False)
        check_result(nums,ans,maxv[0])
        return maxv[0]

############################################################
# Nothing can be changed in check_result
# Note check_result is a global hanging function
###########################################################  
def check_result(a:'Python list',ans:'Python List',amax:'int'):
    x = sorted(ans)
    t = 0
    for e in x:
        t = t + a[e]
    assert(t == amax)
    # assert you did not break the rule
    lx = len(x)
    for i in range(0,lx-1,2):
        pos1 = x[i]
        pos2 = x[i +1]
        assert(pos2 >= (pos1+1))
 
```
